```python
from bs4 import BeautifulSoup
import requests
# from selenium import webdriver
# from selenium.webdriver.common.by import By

import pandas as pd
```

```python
url = "https://www.google.com/finance/markets/climate-leaders"

response = requests.get(url)

soup = BeautifulSoup(response.content)
```

```python
description_CDP= 'Climate leadersCompanies with a CDP climate score of A for the last 3 years (2019 - 2021)'
# if description_CDP
```

```python
for item in data:
    company_info = item.text.strip().split('add_circle_outline')
    
    description_CDP= 'Climate leadersCompanies with a CDP climate score of A for the last 3 years (2019 - 2021)'
    if description_CDP in company_info[0]:
        company_info[0] = company_info[0].replace(description_CDP, '')

    for info in company_info:
        # Splitting by the currency symbols
        currencies = ['', '', '$', 'GBX\xa0', 'CHF\xa0', 'kr\xa0']
        split_info = None

        for currency in currencies:
            if currency in info:
                split_info = info.split(currency)
                break

        if split_info:
            # print(split_info)
            # Extracting company name
            company_name = split_info[0]

            # Extracting price, augmentation, and percentage
            price = split_info[1].replace('+', '').replace('-', '').replace(',', '')
            sign_info = split_info[1][-1] # + or -
            
            augmentation = split_info[2].split('%')[0].split('.')[0]+'.' + split_info[2].split('%')[0].split('.')[1][:2] 
            # percentages= split_info[2].split('%')[0]
            
            print(f"Company: {company_name}, Price: {currency}{price}, Augmentation: {currency}{augmentation}")
```

```python
company_list = []
price_list = []
augmentation_list = []
sign_info_list = []
sign_bool_list =[]

for item in data:
    company_info = item.text.strip().split('add_circle_outline')
    
    description_CDP= 'Climate leadersCompanies with a CDP climate score of A for the last 3 years (2019 - 2021)'
    if description_CDP in company_info[0]:
        company_info[0] = company_info[0].replace(description_CDP, '')

    for info in company_info:
        # Splitting by the currency symbols
        currencies = ['', '', '$', 'GBX\xa0', 'CHF\xa0', 'kr\xa0']
        split_info = None

        for currency in currencies:
            if currency in info:
                split_info = info.split(currency)
                break

        if split_info:
            # print(split_info)
            # Extracting company name
            company_name = split_info[0]

            # Extracting price, augmentation, and percentage
            price = split_info[1].replace('+', '').replace('-', '').replace(',', '')
            sign = split_info[1][-1] # + or -
            
            augmentation = split_info[2].split('%')[0].split('.')[0]+'.' + split_info[2].split('%')[0].split('.')[1][:2] 
            # percentages= 
            
            print(f"Company: {company_name}, Price: {currency}{price}, Augmentation: {currency}{augmentation}")
            # Append extracted data to respective lists
            company_list.append(company_name)
            price_list.append(f"{currency}{price}")
            augmentation_list.append(f"{currency}{augmentation}")
            sign_bool_list.append(1 if sign == '+' else 0)  # Convert sign to boolean (1 for '+' and 0 for '-')
            sign_info_list.append(sign)

# Create a DataFrame
df = pd.DataFrame({
    'Company': company_list,
    'Price': price_list,
    'Augmentation': augmentation_list,
    'sign_bool_list' : sign_bool_list,
    'Sign_Info': sign_info_list  # Add the sign_info column
})
```
